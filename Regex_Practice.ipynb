```python
import pandas as pd
pd.options.display.max_colwidth = 300
```

```python
df = pd.read_csv("Data/String_Cleaning.csv", header = None)
df
```

```python
import re
```

## Phone Number

```python
phone = '([(]?[0-9]{3}[).-]?[0-9]{3}[.-]?[0-9]{4})'
df['phone_number']= df[1].str.extract(phone)
df
```

## State

```python
state = '([A-Z]{2})'
df['state'] = df[1].str.extract(state)
df
```

## City

```python
city = '([A-Z]{1}[a-z]+[\s]?[A-Za-z]+)'
df['city'] = df[1].str.extract(city)
df
```

## URL

```python
url = '(http[s]?://[\S]+)'
df['url'] = df[1].str.extract(url)
df
```

```python
df = df.drop(columns = 1)
df
```

```python
df = df.rename(columns = {0:'name'})
df
```
